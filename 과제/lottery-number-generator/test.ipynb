```python
import mysql.connector

# 데이터베이스 연결 설정
db_connection = mysql.connector.connect(
    host='localhost',        # MySQL 서버 주소
    user='root',    # MySQL 사용자명
    password='0252', # MySQL 비밀번호
    database='lotto'  # 연결할 데이터베이스 이름
)

# 커서 생성
cursor = db_connection.cursor()

# SQL 쿼리 실행
cursor.execute("SELECT * FROM lotto")

# 결과 가져오기
results = cursor.fetchall()

# 결과 출력
for row in results:
    print(row)
```

```python
# 커서와 연결 종료
cursor.close()
db_connection.close()
```

```python
from flask import Flask, jsonify
from flask_cors import CORS
import threading

app = Flask(__name__)
CORS(app)  # CORS 설정

@app.route('/get-variable')
def get_variable():
    aaa = "와 ai가 번호를 주면 출력이 가능하다고?"
    return jsonify(aaa=aaa)

def run_app():
    app.run(port=5001)  # 원하는 포트로 변경 가능

# Flask 앱을 백그라운드에서 실행
threading.Thread(target=run_app).start()
```
